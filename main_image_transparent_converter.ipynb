| Transparent Background Creator | Convert logos or images with white backgrounds to transparent |
|-|-|

```python
# libs
from PIL import Image # pip3 install Pillow library -> converts image to transparent
from IPython.display import Image as IPythonImage # prints the output in a jupyter notebook -> not necessary but seems good for COMPLETENESS


def make_background_transparent(image_path, output_path):
    image = Image.open(image_path)
    image = image.convert("RGBA")
    
    width, height = image.size
    for x in range(width):
        for y in range(height):
            pixel = image.getpixel((x, y))
            if pixel[0] == 255 and pixel[1] == 255 and pixel[2] == 255: # if pixel is white
                image.putpixel((x, y), (pixel[0], pixel[1], pixel[2], 0)) # set alpha to 0 (transparent)
    
    image.save(output_path, "PNG")

input_image_path = "infocenter_white_background.png"  # path to input file with white background (or just file name if in projectf folder)
output_image_path = "infocenter_transparent.png"  # set the name for the output file

# display the original image
print("Original Image:")
display(IPythonImage(filename=input_image_path))

# make the background transparent and save the result to a new file
make_background_transparent(input_image_path, output_image_path)

# display the image with a transparent background
print("Image with Transparent Background:")
display(IPythonImage(filename=output_image_path))
```
